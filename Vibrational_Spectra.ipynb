```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.fft import fft, fftfreq
```

```python
data = np.loadtxt('vac.xvg', comments=['#','@','&'])
```

```python
data.shape
```

```python
plt.plot(data[:,0][:1000], data[:,1][:1000])
```

```python
spectra = fft(data[:,1])

# Number of sample points
N = data.shape[0]
# sample spacing
T = (data[1,0] - data[0,0]) * 1e-12
xf = fftfreq(N, T) * 3.34e-11
```

```python
plt.plot(xf[:N//10], np.abs(spectra)[0:N//10])
```

```python
import plotly.express as px
```

```python
px.line(x=xf[:N//10], y=np.abs(spectra)[0:N//10])      ## rigid water O-H bond rigid, so no peak at 3400 cm-1, this mode comes from angular fluctuations
```
